```python
import json

import pandas as pd

with open("../data/strapi-pages.json") as f:
    json_data = json.load(f)
```

```python
pages = []
for page in json_data["data"]:
    page_summary = page.get("PageSummary")
    if not page_summary or len(page_summary) < 50:
        # Page summary has not been generated
        continue

    markdown = "\n\n".join([
        chunk["MD"] for chunk in page["Content"] if chunk["MD"]
    ])

    text = "\n\n".join([
        chunk["CleanText"] for chunk in page["Content"]
    ])

    page_dict = {
        "volume": page["Volume"]["Slug"],
        "page": page["Slug"],
        "summary": page["PageSummary"],
        "markdown": markdown,
        "text": text,
    }

    pages.append(page_dict)

df = pd.DataFrame(pages)
df
```

```python
df.to_csv("../data/itell-pages.csv", index=False)
```
